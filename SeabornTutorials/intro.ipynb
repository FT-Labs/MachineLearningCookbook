```python
#Practises using tips dataset
import seaborn as sns
import matplotlib.pyplot as plt
```

If you're searching at 1 feature -> *Univariate Analysis*
If searching 2 features -> *Bivariate Analysis*

```python
df = sns.load_dataset("tips")
df.head()
```

```python
#Using correlation to find interrelation between independent features
#Tips -> Dependent feature
#Others -> Independent features
print(df.dtypes)
df.corr()
sns.heatmap(df.corr())
```

```python
#Univariate analysis

sns.jointplot(x='tip' , y='total_bill' , data=df , kind='scatter')
```

```python
sns.jointplot(x='tip' , y='total_bill' , data=df , kind='hex')
```

```python
sns.jointplot(x='tip' , y='total_bill' , data=df , kind='regression')
```

# **Pair Plot**
- A pair plot is also known as a scatterplot, which one variable in same data row is matched with
another variables value.

```python
#Features need to be integer or floats, other types are excluded.
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

```python
#Dist plots

sns.distplot(df['tip'])
```

```python
#Kde = Kernel density estimation
sns.distplot(df['tip'] , kde = False , bins=15)
```

# Categorical Plots
-> For categorical features, we will use the below plots to visualize them.

- Violin plot
- Count plot
- Bar plot


```python
#Countplots
sns.countplot('smoker' , data=df)
```

```python
sns.countplot('day' , data=df)
```

```python
sns.countplot(y = 'sex' , data=df)
```

```python
#Bar plots
sns.barplot(x='total_bill' , y = 'sex' , data=df)

```

```python
sns.barplot(x = 'smoker' , y = 'total_bill' , data=df)
```

```python
#Using boxplots
sns.boxplot('day' , 'total_bill' , data=df)
```

```python
sns.boxplot(x='total_bill' , y='day' , hue='sex' , data=df)
```

```python
#Violin Plot
#It helps us to see both the distribution of data in terms of kernel density estimation and the box plot
sns.violinplot(x='total_bill' , y='day' , data=df , palette='rainbow ')
```
