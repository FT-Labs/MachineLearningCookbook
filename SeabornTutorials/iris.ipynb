```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

params = {
    'axes.labelsize': "small",
    'axes.labelcolor' : 'black',
    'xtick.labelsize': 'x-small',
    'xtick.color' : 'black',
    'ytick.color' : 'black',

    'axes.titlecolor' : 'black',
    'text.color' : 'black',
    'figure.figsize' : [15,6],
}
print(plt.rcParams.keys())
plt.rcParams.update(params)
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.corr()
sns.heatmap(df.corr())
```

```python
sns.countplot('species' , data=df)
```

```python
sns.jointplot(x='sepal_length' , y = 'sepal_width' , kind='hex' , data = df)
```

```python
sns.jointplot(x='petal_length' , y='petal_width' , data=df)
```

```python
sns.jointplot(x='petal_length' , y='petal_width' , data=df , kind='regression')

#As can be seen from the graph, there is almost a linear relationship between petal_length and petal_width.
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df , hue='species')
```

```python
sns.barplot(x='petal_length' , y='petal_width' , data=df , hue='species' )
```

```python
sns.barplot('sepal_length' , 'sepal_width' , data=df , hue='species')
```

```python
plt.subplot(2,2,1)
sns.boxplot(x='sepal_length' , y='species' , data=df)
plt.subplot(2,2,2)
sns.boxplot(x='sepal_width' , y='species' , data=df)
plt.subplot(2,2,3)
sns.boxplot(x='petal_length' , y='species' , data=df)
plt.subplot(2,2,4)
sns.boxplot(x='petal_width' , y='species' , data=df)
```
